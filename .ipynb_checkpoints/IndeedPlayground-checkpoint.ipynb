```python
someStr = jobSummary[0].strip('\n').strip()

someStr
```

```python
def clean_summary(arr):
    new_str = ""
    if len(arr) <= 1:
        clean_str = arr[0].strip('\n').strip()                        
        new_str += clean_str
        return new_str
    for s in arr:
        try:
            clean_str = s.strip('\n').strip()                        
            new_str += clean_str
        except:
            new_str += s.contents[0].string
    return new_str
   
```

```python
someStr = clean_summary(jobSummary)
```

```python
someStr
```

```python
for s in jobSummary:
    print(s.string.replace("   ", ""))
```

```python
jobList = []
for i, j in enumerate(jobDivs):
    newJob = {}
    newJobSummary = jobDivs[i].find_all('span', class_='summary')[0].contents
    newJob['id'] = i
    newJob['summary'] = clean_summary(newJobSummary)
    newJob['title'] = jobDivs[i].find_all('a')[0]['title']
    newJob['link'] = 'https://indeed.com' + jobDivs[0].find_all('a')[0]['href']
    
    jobList.append(newJob)
    
jobList    
```

```python
for i in jobList:
    print(i['title'])
    print("*******************")
    print(i['summary'])
    print("====================")
```
